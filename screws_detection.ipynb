```python
from ultralytics import YOLO
import os

model = YOLO("/content/drive/MyDrive/Case Study/runs/detect/train7/weights/best.pt")

val_folder = "/content/drive/MyDrive/Case Study/val/images"

counts = {}

for img in os.listdir(val_folder):
    img_path = os.path.join(val_folder, img)
    results = model(img_path)

    for r in results:
        for box in r.boxes:
            class_id = int(box.cls[0])
            class_name = model.names[class_id]

            if class_name not in counts:
                counts[class_name] = 0

            counts[class_name] += 1

total = sum(counts.values())

print("Validation Results:")
for k, v in counts.items():
    print(f"{k}: {v}")

print("Total Screws:", total)
```
